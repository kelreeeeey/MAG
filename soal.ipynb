```python
def undak(t, expand=False):
    if t < 0:
        if expand:
            print("syarat t < 0")
        return 0
    if t == 0:
        if expand:
            print("syarat t == 0")
        return 0.5
    else:
        if expand:
            print("syarat t > 0")
        return 1
    
def function_x(t, expand=False):
    if t < 0:
        if expand:
            print("syarat t < 0")
        return 0
    elif 0 <= t < 1:
        if expand:
            print("syarat 0 <= t < 1")
        return t
    elif 1 <= t < 2:
        if expand:
            print("syarat 1 <= t < 2")
        return -t+2
    else:
        if expand:
            print("syarat t > 2")
        return 0
        
    
def impulse(t, expand=False): # h(t) = t * u(t)
    if expand:
        print(f"t*undak(t) = {t}*{undak(t, expand=False)} = {t*undak(t, expand=False)}")
    return t * undak(t, expand=expand)

def gambarin(t, f):
    plt.figure(figsize=(10,5))
    plt.stem(t, f)
    plt.xlabel('titik-titik sample'); plt.ylabel('responnya')
    plt.title('nah, monggo!');
    plt.show()
```

```python
# import numpy library
import numpy as np

# define two 1D arrays
a = xt
b = ht #np.array([5, 6, 7, 8])

# define a function to perform integral convolution
def integral_convolution(x, y):
    # get the length of the arrays
    n = len(x)
    m = len(y)
    # initialize an empty array to store the result
    result = np.zeros(n + m - 1)
    # loop through the result array
    for i in range(n + m - 1):
        # initialize a variable to store the sum of products
        s = 0
        # loop through the x array
        for j in range(n):
            # check if the index is valid for the y array
            if i - j >= 0 and i - j < m:
                # multiply the corresponding elements and add to the sum
                s += x[j] * y[i - j]
        # store the sum in the result array
        result[i] = s
    # return the result array
    return result

# call the function and print the result
print(integral_convolution(a, b))
plt.stem(integral_convolution(a, b))
```
