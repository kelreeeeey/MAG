```python
import numpy as np #library diimport, sepert numpy sebagai np, sehingga kita dapat melakukan perhitungan matematika yang lebih advanced pada environment ini
import matplotlib.pyplot as plt #library matplotlib.pyplot juga diimport sebagai plt agar kita dapat melakukan plottingan berupa grafik


```

```python
### Penyelesaian NO 1

# karena persamaan terdiri dari 2 fungsi sinusoidal maka
# masing-masing memiliki parameter masing
# indeks 1 untuk fungsi cos(t) 
# indeks 2 untuk fungsi sin(t+pi/2)

A1 = 1 # mendefisnisikan amplitudo sinyal 1
A2 = 1 # mendefisnisikan amplitudo sinyal 2

theta1 = 0        # mendefinisikan fase sinyal 1
theta2 = np.pi/2  # mendefinisikan fase sinyal 2

omega1 = 1  # mendefinisikan kecepatan sudut sinyal 1
omega2 = 1  # mendefinisikan kecepatan sudut sinyal 2

T1 = 2*np.pi/omega1 # mendefinisikan periode sinyal 1
T2 = 2*np.pi/omega2 # mendefinisikan periode sinyal 2


# karena pada soal diminta t dimulai dari 0 hingga 2pi
start = 0      # nilai awal t
step = 0.1     # untuk loncatan nilai t
stop = 2*np.pi # batas akhir (namun nilai stop tidak diikutkan)

# maka menggunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari 0 hingga 2pi-step, karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai 0 hingga 2pi
# selisih setiap nilai dalam array t adalah sebesar step

t = np.arange(start, stop+step, step)   # mendefinisikan interval waktu
x = A1*np.cos(omega1*t + theta1) + A2*np.sin(omega2*t + theta2) # mendefinisikan persamaan sinyal

# plotting sinyal
plt.plot(t, x)                     # untuk mendefinisikan sumbu x(=t) dan sumbu y(=x)
plt.xlabel('t')                    # untuk memberikan label pada sumbu x
plt.ylabel('x(t)')                 # untuk memberikan label pada sumbu y
plt.title('Grafik x(t)=cos(t)+sin(t+phi/2)')     # untuk memberikan judul pada grafik
plt.show()                         # untuk menampilakan grafik hasil plottingan



```

```python
### Penyelesaian NO 2

#persamaan merupakan fungsi eksponensial


# karena pada soal diminta t dimulai dari -5 hingga 10
start = -5   # nilai awal t
step = 0.1   # untuk loncatan nilai t
stop = 5     # batas akhir (namun nilai stop tidak diikutkan)

# maka menggunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari -5 hingga 5, karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai -5 hingga 5
# selisih setiap nilai dalam array t adalah sebesar step

t = np.arange(start, stop+step, step) # mendefinisikan interval waktu
x = (1/(2*np.pi)**(1/2))*np.exp((-t**2)/0.4) # mendefinisikan persamaan x(t)

# plotting sinyal
plt.plot(t, x)                     # untuk mendefinisikan sumbu x(=t) dan sumbu y(=x)
plt.xlabel('t')                    # untuk memberikan label pada sumbu x
plt.ylabel('x(t)')                 # untuk memberikan label pada sumbu y
plt.title('Grafik x(t)=(1/(2pi)^0.5)exp^(-t^2/0.4)')     # untuk memberikan judul pada grafik
plt.show()                         # untuk menampilakan grafik hasil plottingan

```

```python
### Penyelesaian NO 3

# persamaan mengandung fungsi step heaviside
# persamaan heaviside akan bernilai 1 jika t>0
# dan bernilai 0 jika t<1

# karena pada soal diminta t dimulai dari -5 hingga 10
start = -5   # nilai awal t
step = 0.1   # untuk loncatan nilai t
stop = 10    # batas akhir (namun nilai stop tidak diikutkan)

# maka menggunakan np.arange, kita mendefisnisikan interval t yang berbentukZ array (kumpulan object dengan jenis yang sama)
# yang dimulai dari -5 hingga 10, karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai -5 hingga 10
# selisih setiap nilai dalam array t adalah sebesar step

t = np.arange(start, stop+step, step) # mendefinisikan interval waktu
H = [1 if t[i]>-2.1 else 0 for i in range(len(t))] # mendefinisikan H. untuk setiap nilai yang berada pada array t yang > 0 maka H akan bernilai 1 jika tidak, maka H akan bernilai 0

x = H*np.exp(-t) # mendefinisikan persamaan x(t)

# plotting sinyal
plt.plot(t, x)                     # untuk mendefinisikan sumbu x(=t) dan sumbu y(=x)
plt.xlabel('t')                    # untuk memberikan label pada sumbu x
plt.ylabel('x(t)')                 # untuk memberikan label pada sumbu y
plt.title('Grafik x(t)=exp^(-t)H(t+2)')     # untuk memberikan judul pada grafik
plt.show()                         # untuk menampilakan grafik hasil plottingan

```

```python
### Penyelesaian NO 4

# persamaan merupakan fungsi Dirac Delta
# persamaan akan bernilai 1 jika t= 0
# dan bernilai 0 jika t=! 0

# interval waktu yang diminta adalah dari -5 hingga 10.
# karena persamaan merupakan fungsi Dirac Delta, maka interval
# waktu dibagi menjadi 3

start1 = -5   # nilai awal interval pertama
step1 = 0.1   # untuk loncatan nilai t
stop1 = 2     # batas akhir interval pertama (namun nilai stop tidak diikutkan)

# digunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari -5 hingga 2, karena fungsi akan bernilai 0 ketika t<2
# sehingga interval pertama berawal dari -5 hingga 2.
# karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai -5 hingga 2
# selisih setiap nilai dalam array t adalah sebesar step

t1 = np.arange(start1,stop1+step1, step1) # mendefinisikan nilai dan interval waktu pertama (-5 hingga 2) yang mana nilai dari persamaan adalah 0
d1 = np.zeros_like(t1) # mendefinisikan array dengan nilai 0  sebanyak nilai pada array t1 karena pada interval ini fungsi akan bernilai =0 untuk setiap nilai pada t2


t2 = [2]       # mendefinisikan interval waktu kedua (t=2)
d2 = [1.0]     # fungsi akan bernilai =1 pada saat t=2


start3 = 2.1   # nilai awal interval ketiga
step3 = 0.1    # loncatan t
stop3 = 10     # nilai akhir interval ketiga (namun nilai stop tidak diikutkan)

# digunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari 2 hingga 0, karena fungsi akan bernilai 0 ketika t>2
# sehingga interval ketiga berawal dari 2 hingga 10.
# karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai -2 hingga 10
# selisih setiap nilai dalam array t adalah sebesar step

t3 = np.arange(start3,stop3+step3, step3) # mendefinisikan interval waktu ketiga  
d3 = np.zeros_like(t3) # mendefinisikan array dengan nilai 0 sebanyak nilai pada array t3 karena pada interval ini fungsi akan bernilai =0 untuk setiap nilai pada t3


d = np.concatenate((d1,d2,d3)) # menggabungkan array d1, d2, da, d3
t = np.concatenate((t1,t2,t3)) # menggabungkan array t1, t2, dan t3

# setelah digabunkan array t dan d dapat diplot ke dalam grafik
# plotting sinyal
plt.plot(t, d)                     # untuk mendefinisikan sumbu x(=t) dan sumbu y(=d)
plt.xlabel('t')                    # untuk memberikan label pada sumbu x
plt.ylabel('x(t)')                 # untuk memberikan label pada sumbu y
plt.title('Grafik x(t)=delta(t-2)')     # untuk memberikan judul pada grafik
plt.show()                         # untuk menampilakan grafik hasil plottingan

```

```python
### Penyelesaian NO 5

# persamaan terdiri dari 2 fungsi Dirac Delta
# persamaan akan bernilai 1 jika t= 0
# dan bernilai 0 jika t=! 0

# interval waktu yang diminta adalah dari -5 hingga 10.
# karena merupakan persamaan memiliki 2 titik kritis (x=1) yaitu t=1 dan t=-2 , maka terdapat 5 interval 2 waktu yaitu;
# 2 interval diantara batas dan titik kritis, 1 interval diantara titik kritis, dan 2 pada masing masing titik kritis.

start1 = -5   # nilai awal intervall pertama
step1 = 0.1   # untuk loncatan nilai t
stop1 = -2     # batas akhir interval pertama (namun nilai stop tidak diikutkan)

# digunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari -5 hingga -2, karena fungsi akan bernilai 0 ketika -5<t<-2
# sehingga interval pertama berawal dari -5 hingga 2.
# karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai -5 hingga 2
# selisih setiap nilai dalam array t adalah sebesar step

t1 = np.arange(start1,stop1+step1, step1) ### mendefinisikan nilai dan interval waktu pertama (-5<t<-2) yang mana nilai dari persamaan adalah 0
d1 = np.zeros_like(t1) # mendefinisikan array dengan nilai 0 sebanyak nilai pada array t1 karena pada interval ini fungsi akan bernilai =0 untuk setiap nilai pada t1


t2 = [-2]      ### mendefinisikan interval waktu kedua (t=-2)
d2 = [1.0]     # fungsi akan bernilai =1 pada saat t=-2


start3 = -1.9   # nilai awal interval ketiga
step3 = 0.1    # loncatan t
stop3 = 1      # nilai akhir interval ketiga (namun nilai stop tidak diikutkan)

# digunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari -2.1 hingga 1, karena fungsi akan bernilai 0 ketika -2.1<t<1
# sehingga interval ketiga berawal dari -1.9 hingga 1.
# karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai -1.9 hingga 1
# selisih setiap nilai dalam array t adalah sebesar step

t3 = np.arange(start3,stop3+step3, step3) ### mendefinisikan interval waktu ketiga  
d3 = np.zeros_like(t3) # mendefinisikan array dengan nilai 0 sebanyak nilai pada array t3 karena pada interval ini fungsi akan bernilai =0 untuk setiap nilai pada t3


t4 = [1]       ### mendefinisikan interval waktu keempat  (t=1)
d4 = [1.0]     # fungsi akan bernilai =1 pada saat t=1


start5 = 1.1   # nilai awal interval kelima
step5 = 0.1    # loncatan t
stop5 = 10     # nilai akhir interval kelima (namun nilai stop tidak diikutkan)

# digunakan np.arange, kita mendefisnisikan interval t yang berbentuk array (kumpulan object dengan jenis yang sama)
# yang dimulai dari 1.1 hingga 10, karena fungsi akan bernilai 0 ketika 1<t<10
# sehingga interval kelima berawal dari 1 hingga 10.
# karena nilai stop tidak diikutkan di dalam array ini, maka nilai stop pada script di bawah
# ditambahkan dengan nilai step sehingga interval t bernilai  hingga 10
# selisih setiap nilai dalam array t adalah sebesar step

t5 = np.arange(start5,stop5+step5, step5) ### mendefinisikan interval waktu kelima  
d5 = np.zeros_like(t5) # mendefinisikan array dengan nilai 0 sebanyak nilai pada array t3 karena pada interval ini fungsi akan bernilai =0 untuk setiap nilai pada t3


d = np.concatenate((d1,d2,d3,d4,d5)) # menggabungkan array d1, d2, d3, d4, dan d5
t = np.concatenate((t1,t2,t3,t4,t5)) # menggabungkan array t1, t2, t3, t4, dan t5

# setelah digabunkan array t dan d dapat diplot ke dalam grafik
# plotting sinyal
plt.plot(t, d)                     # untuk mendefinisikan sumbu x(=t) dan sumbu y(=d)
plt.xlabel('t')                    # untuk memberikan label pada sumbu x
plt.ylabel('x(t)')                 # untuk memberikan label pada sumbu y
plt.title('Grafik x(t)=delta(t-1)+delta(t+2)')     # untuk memberikan judul pada grafik
plt.show()                         # untuk menampilakan grafik hasil plottingan

```
