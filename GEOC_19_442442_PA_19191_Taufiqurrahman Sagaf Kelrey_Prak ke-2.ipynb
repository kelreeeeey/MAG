```python
import numpy as np   # memanggil modul numpy sebagai np
import pylab as plt  # memanggil modul pylab sebagai plt

# Sistem 1, penyelesaian sistem 1

t = np.arange(-5,5+0.1,0.1) # mendefinisikan interval waktu t berupa array dari -5 hingga 5 dengan step 0.1
x11 = t # mendefinisikan input 1 pada sistem 1

y11 = np.sin(x11) # mendefinisikan output 1 sistem 1
y12 = x11 # mendefinisikan output 2 sistem 1

#plot sistem 1

fig,(ax1) = plt.subplots(1, figsize=(7.3,6)) # memetakan fungsi sinyal input sistem 1 , dengan grid 1 baris 1 kolom
ax1.plot(t, x11, label="$x_1(t)$") # memetakan fungsi input 1 sistem 1 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("Input Signal-Sistem 1") # menampilkan judul pada gambar plot input sinyal sistem 1 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal ouput sistem 1 dengan grid 1 baris 2 kolom
ax1.plot(t, y11,'y' ,label="$y_1(t)$") # memetakan fungsi output 1 sistem 1 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("Output 1 Signal-Sistem 1") # memberikan judul pada gambar plot output 1 sinyal sistem 1 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1
ax2.plot(t, y12,'r' , label="$y_2(t)$") # memetakan fungsi output 2 sistem 1 terhadap interval waktu t baris 1 kolom 2
ax2.set_title("Output 2 Signal-Sistem 1") # memberikan judul pada gambar plot output 2 sinyal sistem 1 baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 2

fig,(ax1) = plt.subplots(1, figsize=(7.3,6)) # memetakan fungsi sinyal output sistem 1 dengan grid 1 baris 1 kolom
ax1.plot(t, y11,'y' ,label="$y_1(t)$") # memetakan fungsi output 1 sistem 1 terhadap interval waktu t baris 1 kolom 1
ax1.plot(t, y12,'r' , label="$y_2(t)$") # memetakan fungsi output 2 sistem 1 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("output Signal-Sistem 1") # menampilkan judul pada gambar plot output sinyal sistem 1 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1

plt.show() # menampilkan gambar plot
print(len(t))
```

```python
# Sistem 2, penyelesaian sistem 2

x21 = y11 # mendefinisikan input 1 pada sistem 2 yang merupakan output 1 sistem 1
x22 = y12 # mendefinisikan input 2 pada sistem 2 yang merupakan output 2 sistem 1

y21 = x21*x22  # mendefinisikan output 1 pada sistem 2
y22 = -x21*x22 # mendefinisikan output 2 pada sistem 2

#plot sistem 2

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal input sistem 2 dengan grid 1 baris 2 kolom
ax1.plot(t, x21,'y', label="$x_1(t)$") # memetakan fungsi input 1 sistem 2 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("Input 1 Signal-Sistem 2") # memberikan judul pada gambar plot input 1 sinyal 1 sistem 2 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1
ax2.plot(t, x22, 'r',label="$x_2(t)$") # memetakan fungsi input 2 sistem 2 terhadap interval waktu t baris 1 kolom 2
ax2.set_title("Input 2 Signal-Sistem 2") # memberikan judul pada gambar plot input 2 sinyal 2 sistem 2 baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 2

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal output sistem 2 dengan grid 1 baris 2 kolom
ax1.plot(t, y21, label="$y_1(t)$") # memetakan fungsi output 1 sistem 2 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("Output 1 Signal-Sistem 2") # memberikan judul pada gambar plot output sinyal 1 sistem 2 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1
ax2.plot(t, y22, 'g',label="$y_2(t)$") # memetakan fungsi output 2 sistem 2 terhadap interval waktu t baris 1 kolom 2
ax2.set_title("Output 2 Signal-Sistem 2") # memberikan judul pada gambar plot output sinyal 2 sistem 2 baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 2

fig,(ax1) = plt.subplots(1, figsize=(7.3,6)) # memetakan fungsi sinyal output sistem 2 dalam satu gambar 1 baris 1 kolom
ax1.plot(t, y21, label="$y_1(t)$") # memetakan fungsi output 1 sistem 2 terhadap interval waktu t
ax1.plot(t, y22, 'g',label="$y_2(t)$") # memetakan fungsi output 2 sistem 2 terhadap interval waktu t 
ax1.set_title("Output Signal-Sistem 2") # menampilkan judul pada gambar plot output sinyal sistem 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva

plt.show() # menampilkan gambar plot
```

```python
# Sistem 3, penyelesaian sistem 3

x31 = y21 # mendefinisikan input 1 pada sistem 3 yang merupakan output 1 sistem 2
x32 = y22 # mendefinisikan input 2 pada sistem 3 yang merupakan output 2 sistem 2

a = x31[50:101] # mendefinisikan output sistem 3 UNTUK SYARAT t>=0
b = x32[0:50] # mendefinisikan output sistem 3 UNTUK SYARAT t<0

y31 = np.concatenate((b,a)) # mendefinisikan output untuk sistem 3 dengan menggabungkan array a dan b

#plot sistem 3

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal input sistem 2 dengan grid 1 baris 2 kolom
ax1.plot(t, x31, label="$x_1(t)$") # memetakan fungsi input 1 sistem 3 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("Input 1 Signal-Sistem 3") # memberikan judul pada gambar plot input sinyal 1 sistem 3 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1
ax2.plot(t, x32, 'g', label="$x_2(t)$") # memetakan fungsi input 2 sistem 3 terhadap interval waktu t baris 1 kolom 2
ax2.set_title("Input 2 Signal-Sistem 3") # memberikan judul pada gambar plot input sinyal 2 sistem 3 baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 2

fig,(ax1) = plt.subplots(1, figsize=(7.3,6)) # memetakan fungsi sinyal outpu sistem 2 dengan grid 1 baris 1 kolom
ax1.plot(t, y31, 'k', label="$y_3(t)$") # memetakan fungsi output sistem 3 terhadap interval waktu t baris 1 kolom 1
ax1.set_title("Output Signal-Sistem 3") # memberikan judul pada gambar plot output sinyal sistem 3 baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan keterangan kurva baris 1 kolom 1

plt.show() # menampilkan gambar plot
```

```python
#Exercise convolusi 1

dt = 0.01 # mendefinisikan step fungsi np.arange
tc = np.arange(0,5+dt,dt) # mendefinisikan interval waktu sinyal input dan impuls berupa array dimensi 1 untuk fungsi sinyal input dan sinyal impuls
H = [1 if 0<=tc[i]<=5 else 0 for i in range(len(tc))] # mendefinisikan fungsi Heaviside, step up pada x=0 dan step down pada x=5

x = np.cos(2*np.pi*tc)*H # mendefinisikan fungsi untuk sinyal input
h = np.sin(3*np.pi*tc)*H # mendefinisikan fungsi untuk sinyal impuls

tcy = np.arange(0,10+dt,dt) # mendefinisikan interval waktu sinyal output hasil convolusi berupa array dimensi 1
y = np.convolve(x,h) # mendefinisikan y sebagai operasi convolusi sinyal input dan sinyal impuls

# plot convolute 1

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal input dan impuls dengan grid 1 baris 2 kolom
ax1.plot(tc,x,  label="$x(t)$") # memetakan fungsi sinyal input sumbu x=tc, sumbu y=x
ax1.set_title("sinyal input") # menampilkan judul pada plot baris 1 kolom 1 
ax1.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 1 
ax2.plot(tc,h, 'r-.',label="$h(t)$") # memetakan fungsi sinyal input sumbu x=tc, sumbu y=x
ax2.set_title("sinyal impuls") # menampilkan judul pada plot baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 2

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal output dengan grid 1 baris 2 kolom
ax1.plot(tcy,y, label="$y(t)$") # memetakan fungsi sinyal output sumbu x=tcy, sumbu y=y
ax1.set_title("Sinyal output") # menampilkan judul pada plot baris 1 kolom 1 
ax1.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 1 
ax2.plot(tcy,y*dt, label="$y_n(t)$" ) # memetakan fungsi sinyal input ternormalize sumbu x=tcy, sumbu y=y*dt
ax2.set_title("Sinyal output normalize") # menampilkan judul pada plot baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 2

plt.show() # menampilkan plot
```

```python
#Exercise convolusi 2

dt2 = 0.01 # mendefinisikan step fungsi np.arange
tc2 = np.arange(-3.5,5,dt2) # mendefinisikan interval waktu pertama sinyal input dan impuls berupa array dimensi 1 untuk fungsi sinyal input dan sinyal impuls
tc3 = np.arange(0,5,dt2) # mendefinisikan interval waktu ke empat sinyal input dan impuls berupa array dimensi 1 untuk fungsi sinyal input dan sinyal impuls
tc4 = np.arange(-3.5,0,dt2) # mendefinisikan interval waktu ketiga sinyal input dan impuls berupa array dimensi 1 untuk fungsi sinyal input

H2 = [1 if -2<=tc2[i]<=3 else 0 for i in range(len(tc2))] # mendefinisikan syarat fungsi heaviside untuk sinyal input
H3 = [1 if tc3[i]>0 else 0 for i in range(len(tc3))] # mendefinisikan syarat fungsi heaviside untuk sinyal impuls
 
x2 = np.cos(np.pi*tc2)*H2 # mendefinisikan fungsi sinyal input pada interval waktu pertama berupa array
h2 = (5*np.exp(-tc3))*H3 # mendefinisikan fungsi sinyal impuls pada interval kedua berupa array
h3 = np.zeros_like(tc4) # mendefinisikan fungsi sinyal impuls pada interval ketiga berupa array 0
h4 = np.concatenate((h3,h2)) # mendefinisikan fungsi sinyal impuls dengan menggabukan h3 dan h2


tcy2 = np.arange(-3.5,5,dt) # mendefinisikan interval waktu sinyal output hasil convolusi berupa array dimensi 1
y2 = np.convolve(x2,h4, 'same') # mendefinisikan y sebagai operasi convolusi sinyal input dan sinyal impuls dengan dimensi array=dimensi x2

# plot sinyal x(t), dan h(t)

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal input dan impuls dengan grid 1 baris 2 kolom
ax1.plot(tc2,x2,  label="$x(t)$") # memetakan fungsi sinyal input sumbu x=tc2, sumbu y=x2
ax1.set_title("signal input") # menampilkan judul pada plot baris 1 kolom 1 
ax1.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 1 
ax2.plot(tc2,h4, 'r-.',label="$h(t)$") # memetakan fungsi sinyal impuls sumbu x=tc2, sumbu y=h4
ax2.set_title("signal  impuls") # menampilkan judul pada plot baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 2 

# plot sinyal y(t) 

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal output dengan grid 1 baris 2 kolom
ax1.plot(tcy2,y2, label="$y(t)$") # memetakan fungsi sinyal input sumbu x=tc2, sumbu y=22
ax1.set_title("Sinyal output") # menampilkan judul pada plot baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 1 
ax2.plot(tcy2,(y2*dt), label="$y(t)$") # memetakan fungsi sinyal input sumbu x=tcy2, sumbu y=(y2*dt)
ax2.set_title("Sinyal output normalize") # menampilkan judul pada plot baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 2 

plt.show() # menampilkan plot
```

```python
#Exercise convolusi 3

dt = 0.01 # mendefinisikan step fungsi np.arange
t_1 = np.arange(0,1+dt,dt) # mendefinisikan interval waktu pertama sinyal input dan impuls berupa array dimensi 1 untuk fungsi sinyal input dan sinyal impuls
t_2 = np.arange(1,2+dt,dt) # mendefinisikan interval waktu kedua sinyal input dan impuls berupa array dimensi 1 untuk fungsi sinyal input dan sinyal impuls
t_ = np.concatenate((t_1,t_2)) # mendefinisikan interval waktu gabungan dengan menggabungkan t_1 dengan t_2

x_1 = t_1 # mendefinisikan fungsi sinyal input pada interval waktu pertama berupa array
x_2 = -t_2+2 # mendefinisikan fungsi sinyal input pada interval waktu kedua berupa array
x = np.concatenate((x_1,x_2)) # mendefinisikan fungsi sinyal input dengan menggabungkan x_1 dengan x_2

h1 = 1-t_1 # mendefinisikan fungsi sinyal impuls pada interval waktu pertama berupa array
h2 = np.zeros_like(t_2) # mendefinisikan fungsi sinyal impuls pada interval waktu kedua berupa array 0
h = np.concatenate((h1,h2)) # mendefinisikan fungsi sinyal input dengan menggabungkan h1 dengan h2

ty = np.arange(0,4+(3*dt),dt) # mendefinisikan interval waktu sinyal output hasil convolusi berupa array dimensi 1
y = np.convolve(x,h) # mendefinisikan y sebagai operasi convolusi sinyal input dan sinyal impuls dengan dimensi array=dimensi x+h-1

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal input dan impuls dengan grid 1 baris 2 kolom
ax1.plot(t_,x, label='$x(t)$') # memetakan fungsi sinyal input sumbu x=t_, sumbu y=x
ax1.set_title("Sinyal input") # menampilkan judul pada plot baris 1 kolom 1 
ax1.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 1 
ax2.plot(t_,h,'r-.',label='$h(t)$') # memetakan fungsi sinyal input sumbu x=t_, sumbu y=h
ax2.set_title("Sinyal impuls") # menampilkan judul pada plot baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 2

fig,(ax1, ax2) = plt.subplots(1,2, figsize=(16,6)) # memetakan fungsi sinyal input dan impuls dengan grid 1 baris 2 kolom
ax1.plot(ty,y, label='$y(t)$') # memetakan fungsi sinyal input sumbu x=ty, sumbu y=y
ax1.set_title("Sinyal output") # menampilkan judul pada plot baris 1 kolom 1
ax1.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 1 
ax2.plot(ty,y*dt, label='$y(t)$') # memetakan fungsi sinyal input sumbu x=ty, sumbu y=y*dt
ax2.set_title("Sinyal output normalize") # menampilkan judul pada plot baris 1 kolom 2
ax2.legend(loc='upper left', frameon=True) # menampilkan legenda pada plot baris 1 kolom 2

plt.show()
```
