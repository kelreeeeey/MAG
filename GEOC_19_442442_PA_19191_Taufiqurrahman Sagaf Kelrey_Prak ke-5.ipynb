```python
#import modul
import numpy as np
import soundfile as sf
import pylab as plt
```

Data $real$

```python
#membaca file rekaman
x, fs = sf.read('speech.wav') #import data rekaman suara
t = np.arange(len(x))/fs #mendefinisikan interval waktu
dt = 1/fs

#mendefinisikan R, C dan T
R = 60 # Mendefinisikan nilai resistor
C = 0.5 # Mendefinisikan nilai kapasitor
T = dt # Mendefinisikan periode atau interval sampling
y = np.zeros(x.size)

#mendefinisikan koefisien filter
a0 = (R*C)/T +1
a1 = (R*C)/T
b0 = (R*C)/T
b1 = (R*C)/T

# Keadaan awal (n=0)
y_1 = 0
x_1 = 0
y[0] = (1/a0)*(b0*x[0]+x_1+y_1)

#mendefinisikan persamaan beda
for i in range (1,x.size):
    y[i] = (1/a0)*(b0*x[i]+b1*x[i-1]+a1*y[i-1])
    

#plot data real
plt.figure(figsize=(20,20)) #mengatur ukuran gambar
plt.subplot(4,1,4) #grafik 1
plt.plot(t,x, 'g-') #memetakan gelombang rekaman suara
plt.ylabel('$x$[k]') #memberi label pada sumbu vertikal
plt.xlabel('$t$(s)') #memberi label pada sumbu mendatar
plt.axis([0,t[-1], -1, 1]) #mengatur nilai min. dan maks. sumbu horizontal dan sumbu vertikal
plt.title('Data $Real$ Belum Terolah') #memberi judul pada grafik

plt.figure(figsize=(20,20)) #mengatur ukuran gambar
plt.subplot(4,1,3) #grafik 1
plt.plot(t,y) #memetakan gelombang rekaman suara
plt.ylabel('$x$[k]') #memberi label pada sumbu vertikal
plt.xlabel('$t$(s)') #memberi label pada sumbu mendatar
plt.axis([0,t[-1], -55, 20]) #mengatur nilai min. dan maks. sumbu horizontal dan sumbu vertikal
plt.title('Data $Real$ Terolah') #memberi judul pada grafik

plt.show() #menampilkan grafik
```

Data $sweep$

```python
#Data sintetik
dt = 1/fs # Mendefinisikan dt
t1 = np.arange(0,30+dt,dt) # Mendefinisikan interval waktu t dari 0 hingga 30
x1 = np.sin(-2*50*t1)*np.cos(10+4.2*t1)*np.pi

#mendefinisikan R, C dan T
R = 55
C = 0.5
T = dt
y1 = np.zeros(x1.size)
#mendefinisikan koefisien filter
a0 = (R*C)/T +1
a1 = (R*C)/T
b0 = (R*C)/T
b1 = (R*C)/T

#mendefinisikan initial condition
y_1 = 0
x_1 = 0
y1[0] = (1/a0)*((b0*x1[0]+x_1+y_1))

#mendefinisikan persamaan beda
for i in range (1,x1.size):
    y1[i] = (1/a0)*(b1*x1[i-1]+b0*x1[i]+a1*y1[i-1])

#plot data sintetik
plt.figure(figsize=(25,40))
plt.subplot(4,1,1) #grafik 1
plt.plot(t1,x1, 'g-') # melakukan plot sinyal ke grfafik
plt.title('Data $Sweep$ Belum Terolah') #menampilkan judul grafik
plt.ylabel('$x$[k]') #memberi label pada sumbu vertikal
plt.xlabel('$t$(s)') #menampilkan label di sumbu mendatar

plt.figure(figsize=(25,40))
plt.subplot(4,1,2) #grafik 1
plt.plot(t1,y1, 'b-') # melakukan plot sinyal ke grfafik
plt.title('Data $Sweep$  Terolah') #menampilkan judul grafik
plt.ylabel('$x$[k]') #memberi label pada sumbu vertikal
plt.xlabel('$t$(s)') #menampilkan label di sumbu mendatar

plt.show() #menampilkan grafik
```
