```python
#import modul
import numpy as np
import soundfile as sf
from numpy import cos, sin, pi, absolute, linspace, arange
from scipy.signal import kaiserord, lfilter,lfilter_zi, firwin, freqz, butter, filtfilt
from pylab import figure, clf, plot, xlabel, ylabel, xlim, ylim, title, legend, grid, axes, show
from matplotlib.pyplot import fill_between, subplots_adjust,rcParams
from matplotlib.patches import ConnectionPatch
from matplotlib import axes
```

# Sinyal Random dan Filtering  LPS tipe Butterworth

## Sinyal Random

```python
#membuat sinyal random

#parameter
fs           = 1000
fc           = 100
w            = fc / (fs / 2) #frekuensi kritis

#sinyal random
t            = arange(350)/fs
xa           = np.random.random(t.size)
x            = xa - np.median(xa) #normalisasi

#plot sinyal input
figure(1, figsize=(16,4))
plot(t,x, 'b-', label="Input Signal", linewidth=1, alpha=1)
xlabel("waktu (s)")
ylabel("Amplitudo (m)")
title("Input Signal")

grid(True)
show()
```

### filter LPS orde 4

```python
#membuat filter LPS IIR dan aplikasi pada x

#mendefinisikan filter LPS orde 4
#mendefinisikan koefisien IIR dengan Butterworth
orde          = 4
lebar         = w
b, a          = butter(orde, lebar, output='ba')

#mendefinisikan initial conditon of filter
initial       = lfilter_zi(b,a)

#aplikasi filter digital
y             = filtfilt(b, a, x)

#aplikasi 1x filter pada x
filtered_x1,_ = lfilter(b, a, x, zi=initial*x[0])

#aplikasi 2x filter pada x
filtered_x2,_ = lfilter(b, a, filtered_x1, zi=initial*filtered_x1[0])
```

```python
#plot sinyal input (filtered)
figure(2, figsize=(16,4))
plot(t,y, 'k-', label="Input Signal - Filtered (filtfilt)", linewidth=1.75, alpha=1)
xlabel("Time (s)")
title("Filtered Signal - filtfilt (N=4)")
grid(True)

#plot sinyal input (filtered 1x)
figure(3, figsize=(16,4))
plot(t,filtered_x1, 'g-', label="Input Signal - Filtered 1x (lfilter)", linewidth=1.5, alpha=1)
xlabel("Time (s)")
title("Filtered Signals 1x - lfilter (N=4)")
grid(True)

#plot sinyal input (filtered 2x)
figure(4, figsize=(16,4))
plot(t,filtered_x2, 'r-', label="Input Signal - Filtered 2x (lfilter)", linewidth=1.5, alpha=1)
xlabel("Time (s)")
title("Filtered Signals 2x - lfilter (N=4)")
grid(True)

#plot sinyal terfilter (all)
figure(5, figsize=(16,4))
plot(t,y, 'k-', label="Input Signal - Filtered (filtfilt)", linewidth=1.75, alpha=1)
plot(t,filtered_x1, 'g-', label="Input Signal - Filtered 1x (lfilter)", linewidth=1.5, alpha=0.7)
plot(t,filtered_x2, 'r-', label="Input Signal - Filtered 2x (lfilter)", linewidth=1.5, alpha=0.5)
xlabel("Time (s)")
legend(loc='best', bbox_to_anchor=(0.75, -0.25, 0.5, 0.5))
title("Filtered Signals (N=4)")
grid(True)

#plot sinyal (all)
figure(6, figsize=(16,4))
plot(t,x, 'b-', label="Input Signal", linewidth=1.75, alpha=0.25)
plot(t,y, 'k-', label="Input Signal - Filtered (filtfilt)", linewidth=1.75, alpha=0.75)
plot(t,filtered_x1, 'g-', label="Input Signal - Filtered 1x (lfilter)", linewidth=1.5, alpha=1)
plot(t,filtered_x2, 'r--', label="Input Signal - Filtered 2x (lfilter)", linewidth=1.2, alpha=1)
xlabel("Time (s)")
legend(loc='best', bbox_to_anchor=(0.75, -0.180, 0.5, 0.5))
title("Data (N=4)")
grid(True)


show()
```
