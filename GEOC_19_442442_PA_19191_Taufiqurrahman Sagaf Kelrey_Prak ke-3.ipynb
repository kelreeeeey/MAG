```python
import scipy.signal as signal
from scipy.signal import lsim
import numpy as np
import pylab as plt

pembilang = [1,1,1,1,1,1] # Mendefinisikan array pembilang
penyebut = [1,0,0,0,0,0] # Mendefinisakan array penyebut
x = signal.TransferFunction(pembilang,penyebut)#Mendefinisikan sinyal x(s) 

dt = 0.01 # mendefinisikan nilai dt
t = np.arange(0,5+dt,dt) # Mendefinisikan interval waktu t dari 0 hinga 5
x = signal.impulse(x,T=t) #mengoperasikan nilai transformasi laplace fungsi X  

# Plot sinyal
plt.plot(t,x[1]) #memetakan sinyal
plt.title('Transformasi Laplace Balik Signal') #menampilkan judul grafik
plt.xlabel('waktu (s)') #menampilkan label di sumbu x
plt.ylabel('Amplitudo') #menampilkan label di sumbu y
plt.show() #memunculkan grafik
```

```python
import sympy as sp
t, s = sp.symbols ('t, s')  # mendefinisikan variabel t dan s
z = sp.Heaviside(t) # mendefinisikan fungsi Heaviside H(t) sebagai z
y = t*sp.exp(-t)*z # mendefinisikan fungsi y(t) sebagai y
x = sp.exp(t)*z # mendefinisikan fungsi x(t) sebagai x
h = y/x # menghitung nilai fungsi impulse respons h(t) dan didefinisikan sebagai h
H = sp.laplace_transform(h,t,s) # menghitung nilai fungsi transfer function H(s) dan didefinisikan sebagai H
sp.simplify(H) # menyederhanakan keluaran nilai H
```
