### Laboratory Exercices 2b

```python
#mendefinisikan 3 interval waktu dengan panjang yang sama
x1 = np.arange(0,5,0.05) 
x2 = np.arange(0,5,0.05)
x3 = np.arange(0,5,0.05)

f1 = 0.5 #mendefinisikan nilai frekuensi fungsi cos
f2 = 0.1 #mendefinisikan nilai frekuensi fungsi sin
f3 = 0.7 #mendefinisikan nilai frekuensi fungsi eksponensial kompleks

y1 = np.cos(2*np.pi*f1*x1+np.pi*5/4) #mendefinisikan fungsi cos
y2 = np.sin(2*np.pi*f2*x2) #mendefinisikan fungsi sin
y3 = np.exp((2j*np.pi*x3 + np.pi/5), dtype = "complex_") #mendefinisikan fungsi eksponensial kompleks
y3_real = np.real(np.exp(2j*np.pi*x3 + np.pi/5)) #mendefinisikan bagian riil fungsi eksponensial kompleks
y3_imaginary = np.imag(np.exp(2j*np.pi*x3 + np.pi/5)) #mendefinisikan bagian imaginer fungsi eksponensial kompleks


#menampilkan grafik dalam satu panggilan
plt.figure(figsize=(15,18)) #mengatur ukuran gambar
plt.subplot(4,1,1) #grafik 1
plt.stem(x1,y1, use_line_collection=True, label='f=0.5Hz') #memetakan fungsi diskrit cos
plt.title('Fungsi Diskrit Cos') #memberikan judul
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.legend() #menampilkan keterangan

plt.subplot(4,1,2) #grafik 2
plt.stem(x2,y2, 'C2o',use_line_collection=True, label='f=0.1Hz') #memetakan fungsi diskrit sin
plt.title('Fungsi Diskrit Sin') #memberikan judul
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.legend() #menampilkan keterangan

plt.subplot(4,1,3) #grafik 3
plt.stem(x3,y3, '-', 'C1o',use_line_collection=True, label='f=0.7 Hz') #memetakan fungsi diskrit eksponensial kompleks
plt.title('Fungsi Diskrit Eksponensial Kompleks') #memberikan judul
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.legend() #menampilkan keterangan

plt.subplot(4,1,4) #grafik 4
plt.stem(x3, y3_real, '-', 'C1o', use_line_collection=True, label='Real') #memetakan bagian real fungsi diskrit eksponensial kompleks
plt.stem(x3, y3_imaginary , ':', 'C2o', use_line_collection=True, label='Imaginary') #memetakan bagian imaginer fungsi diskrit eksponensial kompleks
plt.title('Bagian Real dan Imaginer Fungsi Diskrit Eksponensial Kompleks') #memberikan judul
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.xlabel('waktu(s)') #memberi label pada sumbu mendatar
plt.legend() #menampilkan keterangan

plt.show() #menampilkan seluruh grafik
```

```python
#superposisi
y4 = y1+y2+y3 #operasi superposisi fungsi cos, sin, dan eksponensial kompleks

plt.figure(figsize=(18,20)) #mengatur ukuran gambar grafik
plt.subplot(4,1,1) #grafik 5
plt.stem(x1,y4, use_line_collection=True, label='f=0.7Hz') #memetakan fungsi superposisi
plt.title("Fungsi Superposisi") #memberikan judul
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.xlabel('waktu(s)') #memberi label pada sumbu mendatar
plt.show() #menampilkan grafik 5
```

### Laboratory Exercises 2c

```python
#shifting

x1s = x1-1.4 #melakukan shifting ke kiri pada hasil superposisi
y4 = y1+y2+y3 #fungsi superposisi

plt.figure(figsize=(18,26)) #mengatur ukuran gambar grafik
plt.subplot(4,1,1) #grafik 6
plt.stem(x1,y4, use_line_collection=True, label='Superposisi') #memetakan fungsi superposisi
plt.stem(x1s,y4, ':', 'C2o',use_line_collection=True, label='Shifted') #memetakan fungsi superposisi ter-shifted
plt.title('Fungsi Superposisi - $Shifted$')
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.xlabel('waktu(s)') #memberi label pada sumbu mendatar
plt.legend() #menampilkan keterangan

#Flipping

x2s = np.flipud(x1) #melakukan flipping pada hasil superposisi
y4 = y1+y2+y3 #fungsi superposisi

plt.subplot(4,1,2) #grafik  7
plt.stem(x1,y4, 'C2o',use_line_collection=True, label='Superpotition') #memetakan fungsi superposisi
plt.stem(x2s,y4, ':', 'C6o',use_line_collection=True, label='Flipped') #memetakan fungsi superposisi ter-flipped
plt.title('Fungsi Superposisi - $Flipped$')
plt.ylabel('Amplitudo (m)') #memberi label pada sumbu vertikal
plt.xlabel('waktu(s)') #memberi label pada sumbu mendatar
plt.legend() #menampilkan keterangan
plt.show() #menampilkan grafik 6 dan 7
```
